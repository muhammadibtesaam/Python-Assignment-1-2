3. Write a program in which use all the operators we can use in Python

```python
# Write your code here

# Arithmetic Operators
a = 10
b = 3
print("Arithmetic Operators:")
print("a + b =", a + b)  # Addition
print("a - b =", a - b)  # Subtraction
print("a * b =", a * b)  # Multiplication
print("a / b =", a / b)  # Division
print("a % b =", a % b)  # Modulus
print("a ** b =", a ** b)  # Exponentiation
print("a // b =", a // b)  # Floor division
print()

# Assignment Operators
print("Assignment Operators:")
c = 5
print("Initial c =", c)
c += 2
print("c += 2 ->", c)
c -= 1
print("c -= 1 ->", c)
c *= 3
print("c *= 3 ->", c)
c /= 2
print("c /= 2 ->", c)
c %= 2
print("c %= 2 ->", c)
c **= 2
print("c **= 2 ->", c)
c //= 2
print("c //= 2 ->", c)
print()

# Comparison Operators
print("Comparison Operators:")
x = 7
y = 5
print("x == y ->", x == y)  # Equal
print("x != y ->", x != y)  # Not equal
print("x > y ->", x > y)    # Greater than
print("x < y ->", x < y)    # Less than
print("x >= y ->", x >= y)  # Greater than or equal
print("x <= y ->", x <= y)  # Less than or equal
print()

# Logical Operators
print("Logical Operators:")
p = True
q = False
print("p and q ->", p and q)  # Logical AND
print("p or q ->", p or q)    # Logical OR
print("not p ->", not p)      # Logical NOT
print()

# Bitwise Operators
print("Bitwise Operators:")
m = 6  # 110 in binary
n = 3  # 011 in binary
print("m & n ->", m & n)   # AND
print("m | n ->", m | n)   # OR
print("m ^ n ->", m ^ n)   # XOR
print("~m ->", ~m)         # NOT
print("m << 1 ->", m << 1) # Left shift
print("m >> 1 ->", m >> 1) # Right shift
print()

# Membership Operators
print("Membership Operators:")
list1 = [1, 2, 3, 4, 5]
print("3 in list1 ->", 3 in list1)
print("6 not in list1 ->", 6 not in list1)
print()

# Identity Operators
print("Identity Operators:")
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print("a is b ->", a is b)  # False, different objects
print("a is c ->", a is c)  # True, same object
print("a is not b ->", a is not b)
```

4. Completes the following steps of small task:
    - Mention Marks of English , Islamiat and Maths out of 100 in 3 different variables
    - Mention Variable of Total Marks and assign 300 to it
    - Calculate Percentage

```python
# Write your code here


# Step 1: Marks in different subjects
english_marks = 85
islamiat_marks = 90
maths_marks = 95

# Step 2: Total Marks
total_marks = 300

# Step 3: Calculate Obtained Marks
obtained_marks = english_marks + islamiat_marks + maths_marks

# Step 4: Calculate Percentage
percentage = (obtained_marks / total_marks) * 100

# Step 5: Print Results
print("Marks Obtained:")
print("English:", english_marks)
print("Islamiat:", islamiat_marks)
print("Maths:", maths_marks)
print("Total Marks:", total_marks)
print("Percentage:", percentage, "%")
```
